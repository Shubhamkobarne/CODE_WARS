```python
import pandas as pd
import plotly.express as px
```

```python
#Load the dataset using pandas:
data = pd.read_csv('Dataset for Code Wars.csv')
```

```python
data.columns = data.columns.str.replace('\r\n', '')
data.columns = data.columns.str.strip()
```

```python
print(data.columns)
```

```python
#Find the top 10 highest-grossing movies:
data['WorldwideBox Office'] = data['WorldwideBox Office'].str.replace('$', '').str.replace(',', '').astype(float)
top_10_movies = data.nlargest(10, 'WorldwideBox Office')
top_10_movies
```

```python
fig = px.bar(top_10_movies, x='Movie', y='WorldwideBox Office', title='Top 10 Highest Grossing Movies',
             labels={'Movie': 'Movie', 'WorldwideBox Office': 'WorldwideBox Office (in millions)'})
fig.show()

```

```python
#Find the lowest 10 grossing movies:
lowest_10_movies = data.nsmallest(10, 'WorldwideBox Office')
lowest_10_movies
```

```python
#Create a bar graph for the lowest 10 grossing movies:
fig = px.bar(lowest_10_movies, x='Movie', y='WorldwideBox Office', title='Lowest 10 Grossing Movies',
             labels={'Movie': 'Movie', 'WorldwideBox Office': 'WorldwideBox Office (in millions)'})
fig.show()
```

```python
# Convert 'InternationalBox Office' column to numeric type
data['InternationalBox Office'] = pd.to_numeric(data['InternationalBox Office'], errors='coerce')

# Filter based on Released column and retrieve top 10 rows
top_10_international = data[(data['Released'] >= 2012) & (data['Released'] <= 2022)]
top_10_international = top_10_international.nlargest(10, 'InternationalBox Office')
top_10_international
```

```python
fig = px.line(top_10_international, x='Released', y='InternationalBox Office', color='Movie',
              title='Top 10 Highest Grossing Movies Internationally')
fig.show()



```
